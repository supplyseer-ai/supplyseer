```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of inventory levels and demand levels
inventory = np.linspace(0, 50, 20)  # Inventory levels from 0 to 50
demand = np.linspace(0, 30, 20)     # Demand levels from 0 to 30
X, Y = np.meshgrid(demand, inventory)

# Define the changes in inventory and demand as vectors (vector field)
# Assume that inventory decreases with increasing demand, and demand is influenced by inventory level
dY = -0.1 * Y  # Inventory decreases as demand increases
dX = 0.05 * X  # Demand can increase slightly if inventory is available

# Plotting the vector field
plt.figure(figsize=(12, 8))
plt.streamplot(X, Y, dX, dY, color='b', linewidth=1, density=1.5, arrowstyle='->', arrowsize=1.5)
plt.quiver(X, Y, dX, dY, color='r', scale=100, width=0.005, alpha=.5)
plt.title("Vector Field of Inventory Levels vs. Demand Dynamics", fontsize=16)
plt.xlabel("Demand Level")
plt.ylabel("Inventory Level")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Create a grid for inventory and demand
inventory = np.linspace(1, 100, 20)  # Inventory levels from 1 to 100 units
demand = np.linspace(1, 50, 20)      # Demand rates from 1 to 50 units/day
X, Y = np.meshgrid(inventory, demand)

# Calculate stockout probability (example model)
Z = np.exp(-X / (Y + 1))  # Stockout probability is inversely proportional to inventory

# Plotting the 3D surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k', alpha=0.8)

ax.set_title('Stockout Probability as Function of Inventory and Demand', fontsize=16)
ax.set_xlabel('Inventory Level')
ax.set_ylabel('Demand Rate')
ax.set_zlabel('Stockout Probability')

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

def estimate_changes(time_data, inventory_data, demand_data):
    """
    Estimate the rate of change in inventory (dX) and demand (dY) using numerical differentiation.

    :param time_points: List of time points.
    :param inventory_levels: List of inventory levels at each time point.
    :param demand_levels: List of demand levels at each time point.
    :return: Changes in inventory (dX) and demand (dY) as arrays.
    """
    # Ensure arrays are numpy arrays
    inventory_data_arr = np.array(inventory_data)
    demand_data_arr = np.array(demand_data)
    time_data_arr = np.array(time_data)

    # Use np.gradient to calculate the rate of change for inventory and demand over time
    dX = np.gradient(inventory_data_arr, time_data_arr)
    dY = np.gradient(demand_data_arr, time_data_arr)

    return dX, dY

# Example Data (Inventory Levels and Demand over Time)
time_points = [0, 1, 2, 3, 4, 5, 6]  # Time points in days
inventory_levels = [50, 48, 43, 39, 35, 30, 28]  # Inventory levels at each day
demand_levels = [10, 12, 14, 15, 13, 12, 14]  # Demand levels at each day

# Estimate changes using the function
dX, dY = estimate_changes(time_points, inventory_levels, demand_levels)



# Plotting the vector field based on estimated changes at actual data points
plt.figure(figsize=(12, 8))
plt.quiver(demand_levels, inventory_levels, dY, dX, angles='xy', scale_units='xy', scale=1, color='blue')

plt.title("Estimated Change in Inventory and Demand Levels", fontsize=16)
plt.xlabel("Demand Level")
plt.ylabel("Inventory Level")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
demand = np.array([10, 12, 14, 15, 13, 12, 14, 25, 13, 10, 21, 11, 13,])
inventory = [100]

for i in range(len(demand)):
    if inventory[i-1] < 50:
        inventory[i] += 50
    placeholder = inventory[i]

    inventory.append(placeholder - demand[i]) 

# Assuming historical data is available
historical_inventory = inventory[:-1]
historical_demand = demand
time_steps = len(historical_inventory)

# Calculate dY and dX using numerical differences
dY_data = np.diff(historical_inventory)
dX_data = np.diff(historical_demand)

# Create a grid of inventory levels and demand levels for interpolation
inventory_grid = np.linspace(np.min(historical_inventory), np.max(historical_inventory), 20)
demand_grid = np.linspace(np.min(historical_demand), np.max(historical_demand), 20)
X, Y = np.meshgrid(demand_grid, inventory_grid)

# Interpolate dY and dX to create a vector field
dY = griddata((historical_demand[:-1], historical_inventory[:-1]), dY_data, (X, Y), method='linear', fill_value=0)
dX = griddata((historical_demand[:-1], historical_inventory[:-1]), dX_data, (X, Y), method='linear', fill_value=0)

# Plotting the vector field
plt.figure(figsize=(12, 8))
plt.streamplot(X, Y, dX, dY, color='b', linewidth=1, density=1.5, arrowstyle='->', arrowsize=1.5)
plt.title("Vector Field of Inventory Levels vs. Demand Dynamics (Data-Driven)", fontsize=16)
plt.xlabel("Demand Level")
plt.ylabel("Inventory Level")
plt.grid(True)
plt.show()
```
