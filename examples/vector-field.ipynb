```python
demand_data_series = pd.Series([10, 12, 14, 15, 13, 12, 14, 25, 13, 10, 21, 11, 13])

# Create inventory as a lagged function of demand
inventory_data_series = [100]

for i in range(len(demand_data_series)):
    if inventory_data_series[i-1] < 40:
        inventory_data_series[i] += 40
    placeholder = inventory_data_series[i]

    inventory_data_series.append(placeholder - demand_data_series[i]) 

# Assuming historical data is available
historical_inventory = inventory_data_series[:-1]
```

```python
simulation = DemandInventorySimulation(demand_data=demand_data_series, inventory_data=historical_inventory)
simulation.plot_simulation_vector_field()
```

### Example Stockout probability

Assuming the following stockout function: <br>
$Z = e^{-\frac{X}{Y + 1}}$




```python
inventory = np.linspace(1, 100, 20)  # Inventory levels from 1 to 100 units
demand = np.linspace(1, 50, 20)      # Demand rates from 1 to 50 units/day
stockout_prob = StockoutProbability(inventory_data=np.asarray(inventory), demand_data=np.asarray(demand))
stockout_prob.plot_probability_surface()
```
