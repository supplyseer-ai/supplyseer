```python
from src.supplyseer.models.topological.takens import TakensEmbedding
from src.supplyseer.models.statistical.pca import PCA
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Generate example time series
t = np.linspace(0, 10*np.pi, 1000)
X = np.sin(t) + 0.6*np.random.randn(len(t))

```

```python
# Create and apply embedding
embedding = TakensEmbedding(time_delay=1, embedding_dimension=50, normalize=True)
X_embedded = embedding.fit_transform(X)
```

```python
X_embedded.shape
```

```python
pca = PCA(X_embedded)
tda_pca = pca.fit_pca()["pca_data"]
```

```python
import plotly.express as px
```

```python
time_color = np.linspace(0, 1, len(X_embedded))

fig = px.scatter_3d(x=tda_pca[:, 0],
                    y=tda_pca[:, 1],
                    z=tda_pca[:, 2],
                    
                    color=time_color,
                    color_continuous_scale="viridis")
```

```python
fig.show()
```
